```python
#coding:utf-8
'''
author:mapodoufu
'''
import time
import pandas as pd

# 读取数据
part_1 = pd.read_csv('../data/meinian_round1_data_part1_20180408.txt',sep='$')
part_2 = pd.read_csv('../data/meinian_round1_data_part2_20180408.txt',sep='$')
part_1_2 = pd.concat([part_1,part_2])
part_1_2 = pd.DataFrame(part_1_2).sort_values('vid').reset_index(drop=True)
begin_time = time.time()
print('begin')
# 重复数据的拼接操作
def merge_table(df):
    df['field_results'] = df['field_results'].astype(str)
    if df.shape[0] > 1:
        merge_df = " ".join(list(df['field_results']))
    else:
        merge_df = df['field_results'].values[0]
    return merge_df
# 数据简单处理
print('find_is_copy')
print(part_1_2.shape)
is_happen = part_1_2.groupby(['vid','table_id']).size().reset_index()
# 重塑index用来去重
is_happen['new_index'] = is_happen['vid'] + '_' + is_happen['table_id']
is_happen_new = is_happen[is_happen[0]>1]['new_index']

part_1_2['new_index'] = part_1_2['vid'] + '_' + part_1_2['table_id']

unique_part = part_1_2[part_1_2['new_index'].isin(list(is_happen_new))]
unique_part = unique_part.sort_values(['vid','table_id'])
no_unique_part = part_1_2[~part_1_2['new_index'].isin(list(is_happen_new))]
print('begin')
part_1_2_not_unique = unique_part.groupby(['vid','table_id']).apply(merge_table).reset_index()
part_1_2_not_unique.rename(columns={0:'field_results'},inplace=True)
print('xxx')
tmp = pd.concat([part_1_2_not_unique,no_unique_part[['vid','table_id','field_results']]])
# 行列转换
print('finish')
tmp = tmp.pivot(index='vid',values='field_results',columns='table_id')
tmp.to_csv('../data/tmp.csv')
print(tmp.shape)
print('totle time',time.time() - begin_time)
```

# 定义转换半角函数

```python
def strQ2B(ustring):  
    """全角转半角""" 
    ustring = str(ustring)
    rstring = ""  
    for uchar in ustring:  
        inside_code=ord(uchar)  
        if inside_code == 12288:                                            
            inside_code = 32   
        elif (inside_code >= 65281 and inside_code <= 65374):
            inside_code -= 65248  
  
        rstring += chr(inside_code)
    return rstring  
```

# 定义一些辅助清洗的函数

```python
# 判断是或为浮点数
def my_is_float(x):
    value = re.compile(r'^[-+]?[0-9]*[\.]?[0-9]+$')
    result = value.match(str(x))
    if result:
        return 0
    else:
        return 1

# 清洗：'7.42 7.42'
def comp(x):
    x = str(x)
    s = x.split(' ')
    if len(s)==2:
        if (my_is_float(s[0])==0)&(my_is_float(s[1])==0):
            return (float(s[0]) + float(s[1]))/2
        else:
            return x
    else:
        return x

# 清洗：'nan 7.42'   
def comp2(x):
    x = str(x)
    s = x.split(' ')
    if len(s)==2:
        if (s[0]=='nan')&(my_is_float(s[1])==0):
            return s[1] 
        elif (s[1]=='nan')&(my_is_float(s[0])==0):
            return s[0]
    else:
        return x

# 清洗：'7.42 7.42 7.42'  
def comp3(x):
    x = str(x)
    s = x.split(' ')
    if len(s)==3:
        if (my_is_float(s[0])==0)&(my_is_float(s[1])==0)&(my_is_float(s[2])==0):
            return (float(s[0]) + float(s[1]) + float(s[2]))/3
        else:
            return x
    else:
        return x

# 清洗：'nan nan 7.42' 
def comp4(x):
    x = str(x)
    s = x.split(' ')
    rt = 0
    c = 0
    if len(s)==3:
        if my_is_float(s[0])==0:
            rt += float(s[0])
            c += 1
        if my_is_float(s[1])==0:
            rt += float(s[1])
            c += 1
        if my_is_float(s[2])==0:
            rt += float(s[2])
            c += 1
        if c==0:
            return x
        else:
            return (rt/c)
    else:
        return x 
    
# 清洗：'1-2'     
def comp5(x):
    x = str(x)
    s = x.split('-')
    if len(s)==2:
        if (my_is_float(s[0])==0)&(my_is_float(s[1])==0):
            return (float(s[0]) + float(s[1]))/2
        else:
            return x
    else:
        return x

# 清洗：'7.42 7.42 7.42 7.42'  
def comp6(x):
    x = str(x)
    s = x.split(' ')
    if len(s)==4:
        if (my_is_float(s[0])==0)&(my_is_float(s[1])==0)&(my_is_float(s[2])==0)&(my_is_float(s[3])==0):
            return (float(s[0]) + float(s[1])+float(s[2]) + float(s[3]))/4
        else:
            return x
    else:
        return x
    
def comp7(x):
    return str(x).replace('。', '.')
    
def comp_other(x):
    if (x=='弃查')|(x=='未查')|(x=='nan')|(x=='None'):
        return -999
    else:
        return x

# 根据上述函数执行清洗
for i in c_not_count:
    train_merge[i] = train_merge[i].apply(comp)
    train_merge[i] = train_merge[i].apply(comp2)
    train_merge[i] = train_merge[i].apply(comp3)
    train_merge[i] = train_merge[i].apply(comp4)
    train_merge[i] = train_merge[i].apply(comp5)
    train_merge[i] = train_merge[i].apply(comp6)
    train_merge[i] = train_merge[i].apply(comp7)
    train_merge[i] = train_merge[i].apply(comp_other)    
    test_merge[i] = test_merge[i].apply(comp)
    test_merge[i] = test_merge[i].apply(comp2)
    test_merge[i] = test_merge[i].apply(comp3)
    test_merge[i] = test_merge[i].apply(comp4)
    test_merge[i] = test_merge[i].apply(comp5)
    test_merge[i] = test_merge[i].apply(comp6)
    test_merge[i] = test_merge[i].apply(comp7)
    test_merge[i] = test_merge[i].apply(comp_other)
```

```python
# 提取文字中的浮点数
def my_ex_float(x):
    x = str(x)
    temp = re.findall(r'\d+\.?\d*', x)
    if len(temp)==1:
        return temp[0]
    else:
        return x

# 同上，但是针对眼睛度数使用
def my_ex_float2(x):
    dt  = {'0.1': 650, '0.12': 575, '0.15': 500,
           '0.2': 450, '0.25': 400, '0.3': 325,
           '0.4': 250, '0.5': 200, '0.6': 150,
           '0.8': 100, '1.0': 50, '1.2': 25,
           '1.5': 12.5, '2.0': 6}
    x = str(x)
    temp = re.findall(r'\d+\.?\d*', x)
    if len(temp)==2:
        return (dt[temp[0]]+dt[temp[1]])/2
    else:
        return x
```

# 提取特征sugar_high_related以及sex

```python
concat_result = pd.read_csv('../data/tmp2.csv')

def is_sh_related(x):
    x = str(x)
    if ('高血压' in x) | ('血脂' in x) | ('糖尿病' in x) |('妊娠' in x) | ('肝炎' in x) | ('动脉硬化' in x)|('肾病综合症' in x)|('心肌梗死' in x)| ('脂肪肝' in x) | ('胰岛' in x) |('冠心病' in x)|('贫血' in x)|('胰腺炎' in x)|('血压偏高' in x)|('肥胖' in x):
        return 1
    else:
        return 0

def is_sex(x):
    x = str(x)
    if ('阴道' in x)|('子宫' in x)|('妇' in x)|('乳房' in x)|('孕' in x)|('阴道' in x)|('女' in x)|('宫颈' in x)|('妊娠' in x)|('剖腹产' in x)|('外阴' in x):
        return 1
    elif ('前列腺' in x)|('包皮' in x)|('包茎' in x)|('男' in x)|('吸烟' in x):
        return 2
    else:
        return 0
    
concat_result['sugar_high_related'] = concat_result['all_result'].apply(is_sh_related)
concat_result['sex'] = concat_result['all_result'].apply(is_sex)

train_merge = pd.merge(train_merge, concat_result[['vid', 'sugar_high_related', 'sex']], on=['vid'], how='left')
test_merge = pd.merge(test_merge, concat_result[['vid', 'sugar_high_related', 'sex']], on=['vid'], how='left')
```

# 使用xgboost模型进行预测

## 定义两个训练的函数：
- 第一个函数用于线下训练取最优迭代；
- 第二个函数根据最优迭代训练全部数据，函数存在冗余代码，请见谅。

```python
# 线下训练取最优迭代
def train_model_xgb(params, dtrain, dvalid, predictors, target='target', objective='reg:linear', metrics='rmse',
                 feval=None, early_stopping_rounds=20, num_boost_round=3000, verbose_eval=10, categorical_features=None):
    
    xgb_params={
            'booster':'gbtree',
            'objective': objective,
            'eval_metric':metrics,
            'gamma':0.1,
            'min_child_weight':1.1,
            'max_depth':5,
            'lambda':2,
            'subsample':0.7,
            'colsample_bytree':0.7,
            'colsample_bylevel':0.7,
            'eta': 0.01,
            'tree_method':'exact',
            'seed':0,
            'nthread':24,
    }

    xgb_params.update(params)

    print("preparing validation datasets")

    xgtrain = xgb.DMatrix(dtrain[predictors], label=dtrain[target], feature_names=predictors, nthread=24)
    
    xgvalid = xgb.DMatrix(dvalid[predictors], label=dvalid[target], feature_names=predictors, nthread=24)

    evals_results = {}

    watchlist = [(xgtrain,'train'),(xgvalid,'val')]
    bst1 = xgb.train(xgb_params, 
                     xgtrain, 
                     num_boost_round=num_boost_round, 
                     evals=watchlist, 
                     early_stopping_rounds=early_stopping_rounds, 
                     verbose_eval = 10
                    )


    n_estimators = bst1.best_iteration
    print("\nModel Report")
    print("n_estimators : ", n_estimators)
    print(metrics+":", bst1.best_score)

    return bst1

# 训练全部数据
def train_all_data_xgb(params, dtrain, dvalid, predictors, target='target', objective='reg:linear', metrics='rmse',
                 feval=None, early_stopping_rounds=20, num_boost_round=3000, verbose_eval=10, categorical_features=None):
    xgb_params={
            'booster':'gbtree',
            'objective': objective,
            'eval_metric':metrics,
            'gamma':0.1,
            'min_child_weight':1.1,
            'max_depth':5,
            'lambda':2,
            'subsample':0.7,
            'colsample_bytree':0.7,
            'colsample_bylevel':0.7,
            'eta': 0.01,
            'tree_method':'exact',
            'seed':0,
            'nthread':24
    }

    xgb_params.update(params)

    print("preparing validation datasets")

    xgtrain = xgb.DMatrix(dtrain[predictors], label=dtrain[target], feature_names=predictors, nthread=24)
    
    xgvalid = xgb.DMatrix(dvalid[predictors], label=dvalid[target], feature_names=predictors, nthread=24)

    evals_results = {}

    watchlist = [(xgtrain,'train')]
    bst1 = xgb.train(xgb_params, 
                     xgtrain, 
                     num_boost_round=num_boost_round, 
                     evals=watchlist, 
                     verbose_eval = 10
                    )

    print('finish')

    return bst1
```
